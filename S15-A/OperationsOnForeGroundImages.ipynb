### Need to be consolidated into a single folder

```python
import os
import shutil

source = './fg-raw'
target = './fg'

src_files = os.listdir(source)
trgt_files = os.listdir(target)

for file_name in src_files:
    full_file_name = os.path.join(source, file_name)
    if os.path.isfile(full_file_name):
        if file_name not in trgt_files:
            print("The {} is not in target".format(file_name))
            shutil.copy(full_file_name, target)
        
        else:
            print("All the files are copied from source to target")
            break

```

### Masking the images

```python
from PIL import Image, ImageOps
import glob

source_dir_path = './fg/*.jpg'
target_dir_path = './fg-mask/'

for file in glob.glob(source_dir_path):
    filename = file.split('/')[2]
    im = Image.open(file).convert('RGB')
    im_gray_scale = im.convert('L')
    im_invert = ImageOps.invert(im_gray_scale)
    print("Saving image {} into dir {}".format(filename,target_dir_path))
    im_invert.save(target_dir_path+filename)
```

### Flipping images - mirroring 

```python
from PIL import Image, ImageOps
import glob

source_dir_path = './fg/*.jpg'
target_dir_path = './fg-flip/'

for file in glob.glob(source_dir_path):
    filename = file.split('/')[2]
    im = Image.open(file).convert('RGB')
    im_flip = ImageOps.mirror(im)
    print("Saving image {} into dir {}".format(filename,target_dir_path))
    im_flip.save(target_dir_path+filename)
```

```python
import cv2

fg_img_dir = './fg/'
bg_img_dir = './bg/'
# Load two images
img1 = cv2.imread(bg_img_dir+'img_bg_001.jpg')
img2 = cv2.imread(fg_img_dir+'img_fg_001.jpg')

img2 = cv2.resize(img2,(100,100),interpolation=cv2.INTER_AREA)
# I want to put logo on top-left corner, So I create a ROI
rows,cols,channels = img2.shape
roi = img1[0:rows, 0:cols]
# Now create a mask of logo and create its inverse mask also
img2gray = cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
ret, mask = cv2.threshold(img2gray, 10, 255, cv2.THRESH_BINARY)
mask_inv = cv2.bitwise_not(mask)
# Now black-out the area of logo in ROI
img1_bg = cv2.bitwise_and(roi,roi,mask = mask_inv)
# Take only region of logo from logo image.
img2_fg = cv2.bitwise_and(img2,img2,mask = mask)
# Put logo in ROI and modify the main image
dst = cv2.add(img1_bg,img2_fg)
img1[0:rows, 0:cols ] = dst
cv2.imshow('res',img1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Mean and Std Deviation

```python
import glob
from PIL import Image
import numpy as np

root_dir_path = './fg-bg-test/*.png'
sd=[]
mean = []
for filename in glob.glob(root_dir_path):
    img = Image.open(filename)
    imarr = np.array(img,dtype=np.float)
    mean.append(np.mean(imarr))
    sd.append(np.std(imarr))
    


print("Standard Deviation : ",np.std(sd))
print("Mean : ", np.mean(mean))
```
